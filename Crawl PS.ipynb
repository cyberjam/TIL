```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs


# 만만한 문제부터 

result_df  = pd.DataFrame()

for page in range(3):
    req  = requests.get('https://programmers.co.kr/learn/challenges?tab=all_challenges&challenge_levels%5B%5D=2&challenge_languages[]=python3&page='+ str(page+1))
    html = req.text
    soup = bs(html,'html.parser')
    
    titles = soup.select('#tab_all_challenges > section > div > div.challenge__algorithms--index.col-md-8 > div.algorithm-list > div.row > div > div > a > h4')
    fin_cnts = soup.select('#tab_all_challenges > section > div > div.challenge__algorithms--index.col-md-8 > div.algorithm-list > div.row > div > div > a > div > h6.finished-count')
    links = soup.select('#tab_all_challenges > section > div > div.challenge__algorithms--index.col-md-8 > div.algorithm-list > div.row > div > div > a ')
    
    title_list  = [title.text for title in titles]
    cnt_list  = [int(cnt.text[:cnt.text.find("명")]) for cnt in fin_cnts]
    link_list = ['https://programmers.co.kr'+link.get('href') for link in links]
    
    df = pd.DataFrame({'문제 이름':title_list, '완료한 사람':cnt_list, 'URL':link_list})
    result_df= pd.concat([result_df,df])
    
result_df = result_df.sort_values(by='완료한 사람',ascending=False)
```

```python
result_df.to_csv('프로그래머스 문제.csv', encoding='cp949', index=False)
```
